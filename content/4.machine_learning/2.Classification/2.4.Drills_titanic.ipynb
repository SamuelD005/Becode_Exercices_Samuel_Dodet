```python
def cleaning(df_train):
    df_train['Title'] =df_train['Name'].str.extract(pattern_title)
    df_train['Ticket_1'] = df_train['Ticket'].str.findall(pattern_before)
    df_train['Ticket_2'] =df_train['Ticket'].str.findall(pattern_after) 
    for index,row in df_train.iterrows():
        if row['Ticket_2'] == []:
            df_train.loc[index,'Ticket_2'] = row['Ticket_1']
        else:
            df_train.loc[index,'Ticket_2'] = df_train.loc[index,'Ticket_2'][0]
    df_train['Ticket_1']= df_train['Ticket_1'].str[0]
    df_train['class_ticket'] = df_train['Ticket_1'].map(lambda x: re.sub(r'\W+', '', x))
    df_train['class_ticket'] = df_train['class_ticket'].str.replace('\d+', '')
    df_train['Title'] = df_train['Title'].map(lambda x: re.sub(r'\W+', '', x))
    del df_train['Name']
    del df_train['Ticket_1']
    del df_train['Ticket']
    df_train['Cabin'] = df_train['Cabin'].fillna("Unknown")
    df_train['class_ticket'] = df_train['class_ticket'].fillna("Unknown")
    df_train['Age']= df_train['Age'].fillna(df_train['Age'].mean())
    df_train['Embarked'] = df_train['Embarked'].fillna("U")
    df_train['num_ticket'] = df_train['Ticket_2']
    del df_train['Ticket_2']
    df_train['num_ticket'] = df_train['num_ticket'].map(lambda x: re.sub(r'\W+', '', x))
    df_train['num_ticket'] = df_train['num_ticket'].str.replace('\D+', '')
    df_train['num_ticket'].replace('',0, inplace=True)
    df_train.astype({'Sex': 'str','Cabin': 'str','Embarked': 'str','Title': 'str','num_ticket': 'int64'}).dtypes
```
